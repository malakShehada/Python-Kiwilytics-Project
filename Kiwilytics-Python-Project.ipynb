### What is the total revenue after filling in the missing unit prices with the average price for each product?

```python
# filling missing values in unit_price with the mean of each product
df['unit_price'] = df.groupby('product')['unit_price'].transform(lambda x: x.fillna(x.mean()))

df['total_price'] = df['quantity'] * df['unit_price']
total_revenue = df['total_price'].sum()

print("Total Revenue after filling missing prices:", total_revenue)
```

### Which product has the highest total quantity sold?

```python
# group the total quantity by product
quantity_per_prod = df.groupby('product')['quantity'].sum()
#print(quantity_per_prod)
top_product = quantity_per_prod.idxmax()
print("Product with the highest quantity sold is:", top_product)
```

### Which customer has the highest total spending across all orders?

```python
# group the spending by customer_name
spending_per_customer = df.groupby('customer_name')['total_price'].sum()
top_customer = spending_per_customer.idxmax()
print("Customer with the highest spending is:", top_customer)
```
